```python
import json
import pandas as pd


with open('/content/temiz data.txt', 'r') as f:
    raw_data = f.read()

if not raw_data.strip().startswith('['):
    raw_data = "[" + raw_data.strip().rstrip(',') + "]"

data = json.loads(raw_data)

parsed_data = []
for entry in data:
    start = entry.get("startTime")
    end = entry.get("endTime")

    if "visit" in entry:
        place = entry["visit"]["topCandidate"]
        parsed_data.append({
            "type": "visit",
            "startTime": start,
            "endTime": end,
            "semanticType": place.get("semanticType"),
            "placeLocation": place.get("placeLocation"),
            "placeID": place.get("placeID"),
            "probability": place.get("probability")
        })
    elif "activity" in entry:
        act = entry["activity"]
        parsed_data.append({
            "type": "activity",
            "startTime": start,
            "endTime": end,
            "activityType": act["topCandidate"].get("type"),
            "startLocation": act.get("start"),
            "endLocation": act.get("end"),
            "distanceMeters": act.get("distanceMeters")
        })
df = pd.DataFrame(parsed_data)
df.head(5)
```
